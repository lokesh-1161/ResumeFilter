```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
stopwords.words('english')
from nltk.tokenize import word_tokenize
res=[re.sub(r'[^a-zA-Z0-9.-]',' ',string) for string in resume]
print(res)
str1 = " " 
res=str1.join(res)
print(res)
text_tokens = word_tokenize(res)

tokens = [word for word in text_tokens if not word in stopwords.words()]

print(tokens)
filtered_sentence = (" ").join(tokens)
train1=filtered_sentence
```

```python
# Initializie score counters for each area
edu = 0
coll = 0
exper = 0
skill = 0
roll = 0
other = 0

# Create an empty list where the scores will be stored
scores = []
for area in terms.keys():
        
    if area == 'Education':
        for word in terms[area]:
            if word in tokens1:
                edu +=1
            else:
                edu+=0.5
        scores.append(edu)
        
    elif area == 'Role':
        for word in terms[area]:
            if word in tokens1:
                roll +=1
            else:
               roll+=0.5
        scores.append(roll)
        
    elif area == 'Experiance':
        for word in terms[area]:
            if word in tokens1:
              exper+=1
            else:
              exper+=0.5
        scores.append(exper)
        
    elif area == 'SkillSet':
        for word in terms[area]:
            if word in tokens1:
              skill+=1
            else:
              skill+=0.5
        scores.append(skill)
        
    elif area == 'College':
        for word in terms[area]:
            if word in tokens1:
                coll+=1
            else:
               coll+=0.5
        scores.append(coll)
        
    else:
        for word in terms[area]:
            if word in tokens1:
                other +=0.5
        scores.append(other)
```
